```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Obtener los datos del precio del dólar utilizando yfinance
df_petroleo = yf.download("BZ=F", start="2015-01-01", end="2020-12-31")
```

```python
df_petroleo
```

```python
# Obtener los datos del precio del dólar utilizando yfinance
df_dolar = yf.download("EURUSD=X", start="2015-01-01", end="2020-12-31")
```

```python
df_petroleo = df_petroleo.drop(columns=["Open", "High", "Low", "Adj Close", "Volume"])
```

```python
# Precio día anterior
df_petroleo['prev_day_close'] = df_petroleo['Close'].shift(1)

# Precio semana anterior
df_petroleo['prev_week_close'] = df_petroleo['Close'].shift(5)

# Precio mes anterior
df_petroleo['prev_month_close'] = df_petroleo['Close'].shift(20)

# Trimestre anterior
df_petroleo['prev_qtr_close'] = df_petroleo['Close'].shift(60)

df_petroleo['prev_day_close'].fillna(df_petroleo['Close'], inplace=True)
df_petroleo['prev_week_close'].fillna(df_petroleo['Close'], inplace=True)
df_petroleo['prev_month_close'].fillna(df_petroleo['Close'], inplace=True)
df_petroleo['prev_qtr_close'].fillna(df_petroleo['Close'], inplace=True)

df_petroleo
```

```python
df_dolar = df_dolar.drop(columns=["Open", "High", "Low", "Adj Close", "Volume"])
```

```python
# Precio día anterior
df_dolar['prev_day_close'] = df_dolar['Close'].shift(1)

# Precio semana anterior
df_dolar['prev_week_close'] = df_dolar['Close'].shift(5)

# Precio mes anterior
df_dolar['prev_month_close'] = df_dolar['Close'].shift(20)

# Trimestre anterior
df_dolar['prev_qtr_close'] = df_dolar['Close'].shift(60)

df_dolar['prev_day_close'].fillna(df_dolar['Close'], inplace=True)
df_dolar['prev_week_close'].fillna(df_dolar['Close'], inplace=True)
df_dolar['prev_month_close'].fillna(df_dolar['Close'], inplace=True)
df_dolar['prev_qtr_close'].fillna(df_dolar['Close'], inplace=True)

df_dolar
```

```python
# Porcentaje de cambio: capturar patrones de tendencia
df_petroleo['price_change_pct'] = (df_petroleo['prev_day_close'] - df_petroleo['prev_day_close'].shift(1)) / df_petroleo['prev_day_close'].shift(1)
# Porcentaje de cambio semanal y mensual
df_petroleo['weekly_change_pct'] = df_petroleo['prev_day_close'].pct_change(5)
df_petroleo['monthly_change_pct'] = df_petroleo['prev_day_close'].pct_change(20)
df_petroleo['qtr_change_pct'] = df_petroleo['prev_day_close'].pct_change(60)

# Desviación estándar: volatilidad histórica del precio
df_petroleo['close_std'] = df_petroleo['prev_day_close'].rolling(window=7).std()

# Rellena los valores faltantes en las columnas "price_change_pct" y "close_std" con forward fill (ffill) utilizando fillna()
df_petroleo['price_change_pct'].fillna(method='bfill', inplace=True)
df_petroleo['weekly_change_pct'].fillna(method='bfill', inplace=True)
df_petroleo['monthly_change_pct'].fillna(method='bfill', inplace=True)
df_petroleo['qtr_change_pct'].fillna(method='bfill', inplace=True)
df_petroleo['close_std'].fillna(method='bfill', inplace=True)
```

```python
# Porcentaje de cambio: capturar patrones de tendencia
df_dolar['price_change_pct'] = (df_dolar['prev_day_close'] - df_dolar['prev_day_close'].shift(1)) / df_dolar['prev_day_close'].shift(1)
# Porcentaje de cambio semanal y mensual
df_dolar['weekly_change_pct'] = df_dolar['prev_day_close'].pct_change(5)
df_dolar['monthly_change_pct'] = df_dolar['prev_day_close'].pct_change(20)
df_dolar['qtr_change_pct'] = df_dolar['prev_day_close'].pct_change(60)

# Desviación estándar: volatilidad histórica del precio
df_dolar['close_std'] = df_dolar['prev_day_close'].rolling(window=7).std()

# Rellena los valores faltantes en las columnas "price_change_pct" y "close_std" con forward fill (ffill) utilizando fillna()
df_dolar['price_change_pct'].fillna(method='bfill', inplace=True)
df_dolar['weekly_change_pct'].fillna(method='bfill', inplace=True)
df_dolar['monthly_change_pct'].fillna(method='bfill', inplace=True)
df_dolar['qtr_change_pct'].fillna(method='bfill', inplace=True)
df_dolar['close_std'].fillna(method='bfill', inplace=True)
```

```python
df_petroleo.to_csv('datos_precio_petroleo_sineventos.csv')
df_dolar.to_csv('datos_cambio_dolar_sineventos.csv')
```

```python
# Diccionario de eventos actualizado
eventos = {
    "Desaceleración económica en China": ("2014-07-01", "2020-12-31"),
    "Aumento de la producción de petróleo de esquisto en Estados Unidos": ("2005-01-01", "2020-12-31"),
    "Acuerdo nuclear con Irán y sanciones": ("2015-07-14", "2020-12-31"),
    "Sanciones de Estados Unidos": ("2018-05-01", "2020-12-31"),
    "Crisis de la deuda en Europa": ("2009-01-01", "2020-12-31"),
    "Pandemia COVID-19": ("2019-12-31", "2020-12-31")
}

# Agregar información de eventos al DataFrame del precio del petróleo
for evento, fechas in eventos.items():
    fecha_inicio = pd.to_datetime(fechas[0])
    fecha_fin = pd.to_datetime(fechas[1])

    # Crear una columna para indicar la presencia o ausencia del evento en cada fecha
    df_petroleo[evento] = df_petroleo.index.map(
        lambda x: 1 if fecha_inicio <= x <= fecha_fin else 0
    )

    # Crear una columna para indicar la duración del evento en cada fecha
    df_petroleo[f"Duración_{evento}"] = df_petroleo.index.map(
        lambda x: (x - fecha_inicio).days + 1 if fecha_inicio <= x <= fecha_fin else 0
    )

# Agregar información de eventos al DataFrame del tipo de cambio del dólar
for evento, fechas in eventos.items():
    fecha_inicio = pd.to_datetime(fechas[0])
    fecha_fin = pd.to_datetime(fechas[1])

    # Crear una columna para indicar la presencia o ausencia del evento en cada fecha
    df_dolar[evento] = df_dolar.index.map(
        lambda x: 1 if fecha_inicio <= x <= fecha_fin else 0
    )

    # Crear una columna para indicar la duración del evento en cada fecha
    df_dolar[f"Duración_{evento}"] = df_dolar.index.map(
        lambda x: (x - fecha_inicio).days + 1 if fecha_inicio <= x <= fecha_fin else 0
    )
```
